```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
```

```python
def calculate_z(x,mean,std):
    z = (x-mean)/std
    return z
```

```python
def plot_norm_dist(point1, point2=None, mean=0, std=1, label='cdf'):
    # Parameter validation
    if not isinstance(point1, (int, float)):
        raise ValueError("point1 must be a numeric value.")
    if point2 and not isinstance(point2, (int, float)):
        raise ValueError("point2 must be a numeric value.")
    if not isinstance(mean, (int, float)):
        raise ValueError("mean must be a numeric value.")
    if not isinstance(std, (int, float)):
        raise ValueError("std must be a numeric value.")
    if label not in ['cdf', 'sf', 'ppf']:
        raise ValueError("label must be 'cdf', 'sf', or 'ppf'.")

    # Generate x values
    x = np.linspace(mean - 3 * std, mean + 3 * std, num=100)

    # Plotting
    plt.plot(x, stats.norm.pdf(x, loc=mean, scale=std), color='blue', linestyle='-', linewidth=2, label='Normal Distribution')
    plt.fill_between(x, stats.norm.pdf(x, loc=mean, scale=std), color='skyblue', alpha=0.3)  # Fill area under the curve
    plt.axvline(x=mean, ls='--', label='Mean', color='green')

    if label == 'cdf':
        if not point2:
            point2 = -3
        x_fill = np.linspace(point1, point2)
        plt.fill_between(x_fill, stats.norm.pdf(x_fill, loc=mean, scale=std), color='red', alpha=0.3)
        plt.axvline(x=point1, ls='--', label=f'x1={point1}', color='orange')
        plt.axvline(x=point2, ls='--', label=f'x2={point2}', color='orange')
    elif label == 'sf':
        x_fill = np.linspace(point1, mean + 3 * std)
        plt.fill_between(x_fill, stats.norm.pdf(x_fill, loc=mean, scale=std), color='red', alpha=0.3)
        plt.axvline(x=point1, ls='--', label=f'x1={point1}', color='orange')
    elif label == 'ppf':
        percentile = stats.norm.ppf(point1, mean, std)
        print(f'The X value for the {point1*100} percentile is {(percentile):.2f}')
        plt.axvline(x=percentile, ls='--', label=f'x1={point1}', color='orange')

    # Plot customization
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.title('Normal Distribution')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
# plot_norm_dist(1, 2, mean=0, std=1, label='cdf')
# plot_norm_dist(1, mean=0, std=1, label='sf')
# plot_norm_dist(0.95, mean=0, std=1, label='ppf')
```

```python
mu=20
std=0.15

x = np.linspace(mu-3*std,mu+3*std,num=100)
pdf = stats.norm.pdf(x,loc=mu,scale=std)

x_fill = np.linspace(20,20.27)
print(f'The probability of values falling between 20 and 20.27 is {stats.norm.cdf(x_fill[-1],mu,std) - stats.norm.cdf(x_fill[0],mu,std)}')
print(f'The probability of ball bearings having a diameter of 20.27 or moreis {stats.norm.sf(20.27,mu,std)}')
plt.plot(x, pdf, color='blue', linestyle='-', linewidth=2, label='Normal Distribution')
plt.fill_between(x, pdf, color='skyblue', alpha=0.3)  # Fill area under the curve
plt.fill_between(x_fill, stats.norm.pdf(x_fill,loc=mu,scale=std), color='red', alpha=0.3)  # Fill area under the curve
plt.axvline(x=mu,ls='--',label = 'Mean',color='green')
plt.axvline(x=x_fill[0],ls='--',label = f'x1={x_fill[0]}',color='ORANGE')
plt.axvline(x=x_fill[-1],ls='--',label = f'x2={x_fill[-1]}',color='ORANGE')


plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()

###############################################################################################################################################

x_fill = np.linspace(19.85,20.30)
print(f'The probability of values falling between 19.85 and 20.30 is {stats.norm.cdf(x_fill[-1],mu,std) - stats.norm.cdf(x_fill[0],mu,std)}')
print(f'The probability of ball bearings having a diameter of 19.91 or less is {stats.norm.cdf(19.91,mu,std)}')
plt.plot(x, pdf, color='blue', linestyle='-', linewidth=2, label='Normal Distribution')
plt.fill_between(x, pdf, color='skyblue', alpha=0.3)  # Fill area under the curve
plt.fill_between(x_fill, stats.norm.pdf(x_fill,loc=mu,scale=std), color='red', alpha=0.3)  # Fill area under the curve
plt.axvline(x=mu,ls='--',label = 'Mean',color='green')
plt.axvline(x=x_fill[0],ls='--',label = f'x1={x_fill[0]}',color='ORANGE')
plt.axvline(x=x_fill[-1],ls='--',label = f'x2={x_fill[-1]}',color='ORANGE')


plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

#Q23

```python
mu1=180000000
std1=25000000
mu2=1500
std2=120

##clarion 
clarion_pnl=170000000
clarion_empl = 1850
z1 = calculate_z(clarion_pnl,mu1,std1)
z2 = calculate_z(clarion_empl,mu2,std2)

###Plotting the normal distribution
#PnL
x = np.linspace(mu1 - 3*std1,mu1 + 3*std1)
y= stats.norm.pdf(x,loc=mu1,scale=std1)

plt.plot(x,y,color='black',label='Normal Distribution')
plt.fill_between(x,y,color='skyblue')
plt.axvline(mu1,ls='--',label='Mean')
plt.axvline(clarion_pnl,ls='--',label='Clarion Fabrications',color='red')
plt.legend()
plt.show()

#Epmloyees - using the standard normal distribution
mu=0
std=1

x = np.linspace(mu - 3*std,mu + 3*std)
y= stats.norm.pdf(x,loc=mu,scale=std)

plt.plot(x,y,color='black',label='Normal Distribution')
plt.fill_between(x,y,color='skyblue')
plt.axvline(mu,ls='--',label='Mean')
plt.axvline(z2,ls='--',label='Clarion Fabrications',color='red')
plt.legend()
plt.show()

print(f'{stats.norm.sf(z1,0,1)*100:.2f}% of aluminum manufacturers have a higher PnL than Clarion.')
print(f'{stats.norm.sf(z2,0,1)*100:.2f}% of aluminum manufacturers have a higher head count than Clarion.')
```
