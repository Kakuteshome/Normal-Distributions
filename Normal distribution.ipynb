```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
```

```python
def calculate_z(x,mean,std):
    z = (x-mean)/std
    return z
```

```python
print(calculate_z(45,60,5))
print(calculate_z(75,60,5))
```

```python
print(calculate_z(790,1000,100))
```

```python
print(calculate_z(482,400,50))
```

```python
mu=16.5
std=3.5
print('P(z):',stats.norm.cdf(x=20,loc=mu,scale=std) - stats.norm.cdf(x=16.5,loc=mu,scale=std))
print('The percent of the population falling over 20 is', stats.norm.sf(20,mu,std))
print('The percent of the population falling under 15 is', stats.norm.cdf(15,mu,std))

x = np.linspace(mu-3*std,mu+3*std,num=100)
pdf = stats.norm.pdf(x,loc=mu,scale=std)

plt.plot(x, pdf, color='blue', linestyle='-', linewidth=2, label='Normal Distribution')
plt.fill_between(x, pdf, color='skyblue', alpha=0.3)  # Fill area under the curve
plt.axvline(x=mu,ls='--',label = 'Mean',color='green')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
